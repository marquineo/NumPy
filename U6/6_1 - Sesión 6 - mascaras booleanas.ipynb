### Ejercicio 1 — Temperaturas por encima del umbral

**Enunciado:**  
Una estación meteorológica registró las temperaturas diarias (°C) durante dos semanas.  
Filtra y devuelve solo las temperaturas que superan los 30°C (ola de calor).

**Ten en cuenta:**
- Una máscara booleana es un array de `True/False` del mismo tamaño.
- Puedes crear una con `array > valor`.
- Aplica la máscara al array original usando corchetes: `array[mascara]`.


```python
temperaturas = np.array([22.1, 31.5, 28.4, 35.2, 19.8, 33.0, 27.6,
                         30.1, 29.9, 36.4, 21.3, 32.8, 18.5, 34.7])

mask_temp = temperaturas > 30
indices = np.where(mask_temp)[0]
ola_calor = temperaturas[indices]
print(ola_calor)
```

**Variante extra:** Calcula también el promedio de temperaturas durante los días de ola de calor.

**Enunciado:**  
Un profesor tiene las notas finales de 12 estudiantes (escala 0-10).

- `~mascara` invierte una máscara booleana (`True` <-> `False`).

```python
notas = np.array([7.5, 3.2, 9.1, 4.8, 6.0, 2.5, 8.3, 5.0, 1.9, 7.7, 4.1, 6.5])
aporbados = notas >= 5
suspendidos = ~aporbados

# falta hacer np.while()

print(f"aprobados = {aporbados}")
print(f"suspendidos = {suspendidos}")
```

Considera extremo cualquier valor menor de 5 bar **o** mayor de 95 bar. Identifica y cuenta los outliers.

### Ejercicio 7 — Filtro de candidatos en proceso de selección

**Enunciado:**  
Tenemos dos arrays paralelos: puntuaciones técnicas y de entrevista (0-100).  
Un candidato pasa si tiene técnica >= 70 **y** entrevista >= 60,  
**o** si tiene técnica >= 90 (candidato excepcional, pasa directamente).

**Ten en cuenta:**
- Puedes combinar más de dos condiciones: `(A & B) | C`.
- Respeta siempre los paréntesis para controlar la precedencia.


```python
tecnica    = np.array([85, 55, 92, 70, 68, 91, 74, 60, 88, 45, 95, 72])
entrevista = np.array([75, 80, 50, 65, 90, 85, 55, 70, 72, 60, 40, 88])

# mascara
pasa = ((tecnica >= 70) & (entrevista >= 60)) | (tecnica >= 90)

indices = np.where(pasa)[0]

print(tecnica[indices])
```

### Ejercicio 8 — Análisis de edades en un censo

**Enunciado:**  
Un censo recoge edades de habitantes. Separa la población en tres grupos:
- **Menores:** < 18 años  
- **Adultos:** 18 a 64 años (inclusive)  
- **Mayores:** >= 65 años

**Ten en cuenta:**
- Crea tres máscaras separadas, una por grupo.
- Comprueba que son mutuamente excluyentes sumando sus `.sum()`.


```python
edades = np.array([8, 34, 72, 15, 45, 67, 3, 29, 55, 80, 18, 63, 64, 65, 17, 22])

menores = edades < 18
adultos = (edades >= 18) & (edades <= 64)
mayores = edades > 64

index_menores = np.where(menores)[0]
index_adultos = np.where(adultos)[0]
index_mayores = np.where(mayores)[0]

print(edades[index_menores])
print(edades[index_adultos])
print(edades[index_mayores])

assert (menores.sum() + adultos.sum() + mayores.sum() == len(edades))

media_menores = edades[index_menores].sum() / len(edades[index_menores])
media_adultos = edades[index_adultos].sum() / len(edades[index_adultos])
media_mayores = edades[index_mayores].sum() / len(edades[index_mayores])
print(f"media menores = {media_menores}")
print(f"media adultos = {media_adultos}")
print(f"media mayores = {media_mayores}")
```

**Variante extra:** Calcula la edad media de cada grupo usando la máscara correspondiente.

### Ejercicio 9 — Censurar valores extremos (clipping manual)

**Enunciado:**  
Un sensor de temperatura tiene lecturas erróneas: valores por debajo de -20°C o por encima de 60°C son imposibles.  
Reemplaza esos valores con `np.nan` para marcarlos como no válidos.

**Ten en cuenta:**
- `array[mascara] = valor` reemplaza **in place** (modifica el array original).
- `np.nan` solo funciona con arrays de tipo `float`, no `int`.
- Es buena práctica trabajar sobre una copia: `array.copy()`.


```python
temps_sensor = np.array([-25.0, 18.5, 62.3, 30.0, -22.1, 45.6, 15.0,
                          70.0, 28.9, -30.0, 50.2, 55.5], dtype=float)
temps_old = temps_sensor.copy()

fuera_rango = (temps_sensor < -20) | (temps_sensor > 60)

print(f"Lecturas erroneas = {fuera_rango.sum()}")

temps_sensor[fuera_rango] = np.nan
print(temps_sensor)
```

**Variante extra:** En lugar de NaN, reemplaza los valores erróneos con la media de los valores válidos.

### Ejercicio 11 — Normalización condicional de datos de ventas

**Enunciado:**  
Un dataset tiene valores negativos (devoluciones) y positivos (ventas).  
Normaliza **solo** los valores positivos al rango [0, 1] usando la fórmula min-max, dejando los negativos intactos.

**Ten en cuenta:**
- Extrae los valores positivos con la máscara, calcula min y max sobre ellos.
- Fórmula min-max: `(x - min) / (max - min)`.
- Recuerda asignar el resultado de vuelta al array con la misma máscara.


```python
# Configuración para que NO use notación científica y use 4 decimales
np.set_printoptions(suppress=True, precision=4)

datos_ventas = np.array([120.0, -30.0, 450.0, -15.0, 300.0, -80.0,
                         600.0, 80.0, -5.0, 200.0, -50.0, 500.0])

mascara_positivos = datos_ventas >= 0

# Extraemos solo los positivos para calcular sobre ellos
ventas_solo = datos_ventas[mascara_positivos]
v_min = ventas_solo.min()
v_max = ventas_solo.max()

# Aplicamos la normalización y guardamos en el array original
datos_ventas[mascara_positivos] = (ventas_solo - v_min) / (v_max - v_min)

print(datos_ventas)
```

**Variante extra:** Aplica la normalización también a los negativos pero en el rango [-1, 0].

### Ejercicio 12 — Imputación de valores faltantes en encuesta

**Enunciado:**  
Una encuesta de satisfacción (1-5) tiene respuestas faltantes representadas como 0.  
Imputa los valores faltantes con: la mediana de las respuestas válidas.

**Ten en cuenta:**
- Usa la máscara para filtrar los válidos antes de calcular estadísticas.
- Para la moda: `np.bincount(array)` cuenta frecuencias; `np.argmax()` da el índice del máximo.


```python
respuestas = np.array([4, 0, 3, 5, 0, 2, 4, 0, 5, 3, 0, 4, 1, 0, 3])

invalidas = respuestas == 0
# media de respuesta validas
mediana = np.median(respuestas[~invalidas])

print(f"mediana = {mediana}")

# crea una copia de respuestas
encuesta_trucada = respuestas.astype(float)

encuesta_trucada[invalidas] = mediana

print(encuesta_trucada)
```

**Variante extra:** Imputa usando la mediana de los vecinos más cercanos (+-2 posiciones).
