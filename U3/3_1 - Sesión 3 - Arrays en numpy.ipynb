## Atributos de los arrays NumPy

Primero veamos algunos atributos útiles de los arrays. Empezaremos definiendo un array. 

Usaremos el generador de números aleatorios de NumPy, al que le fijaremos una `semilla` para garantizar que se generen los mismos arrays aleatorios cada vez que se ejecute este código:

````python
import numpy as np
np.random.seed(0)  # semilla para reproducibilidad

x1 = np.random.randint(10, size=6)        # Array unidimensional

x1
````

<b>Pregunta:</b> ¿Qué es una semilla?
</div>

<div class="alert alert-block alert-info">

Cada array tiene los atributos `ndim` (número de dimensiones), `shape` (tamaño de cada dimensión) y `size` (tamaño total del array):

````python
print("número de dimensiones de x1: ", x1.ndim)
print("tamaño de cada dimensión de x1:", x1.shape)
print("tamaño total de x1: ", x1.size)
print("tipo del array", print(x1.dtype))
````

<div class="alert alert-block alert-success">
<b>Nota:</b> ndim y shape están más pensados para arrays de varias dimensiones. De momento nosotros no los vamos a usar, así que no nos aportan información útil.
</div>

Otro atributo útil es `dtype`, el tipo de dato del array:

````python
print("dtype:", x1.dtype)
````

Otros atributos son `itemsize`, que indica el tamaño (en bytes) de cada elemento del array, y `nbytes`, que indica el tamaño total (en bytes) del array:

````python
print("itemsize:", x1.itemsize, "bytes")
print("nbytes:", x1.nbytes, "bytes")
````

En general, `nbytes` es igual a `itemsize` multiplicado por `size`.

Resumiendo:

$nbytes = itemsize * size$

- `nbytes`: Tamaño total (en bytes)
- `itemsize`: Tamaño de cada elemento (en bytes)
- `size`: Número de elementos

## Acceso a los Arrays por índice

Los índices en NumPy son como los de Python. 

En un array unidimensional, se puede acceder al valor _i_ (contando desde cero) especificando el índice deseado entre corchetes, <span style="color:Gold">vamos, como siempre</span>:

````python
x1
````

````python
x1[0]
````

````python
x1[4]
````

Para acceder desde el final del array, puedes usar índices negativos:

````python
x1[-1]
````

````python
x1[-2]
````

<div class="alert alert-block alert-info">
<b>Pregunta:</b> ¿Qué muestra x1[-4]?

Ten en cuenta que, a diferencia de las listas de Python, <span style="color:Gold">los arrays de NumPy tienen un tipo fijo</span>. Esto significa, por ejemplo, que si intentas insertar un valor de coma flotante en un array de enteros, el valor se truncará sin avisar.

````python
x1[0] = 3.14159  # Pero el array es de ENTEROS así que lo trunca a 3!
x1
````

## Subarrays: Slicing

Así como podemos usar corchetes para acceder a elementos individuales del array, también podemos usarlos para <span style="color:Gold">acceder a subarrays</span> con la notación conocida como *slice o slicing*, con los dos puntos (`:`).

La sintaxis de slicing de NumPy sigue la de la lista estándar de Python; para acceder a un slice del array `x`, usa:

```python
x[inicio:fin:paso]
```

Si alguno de estos no se especifica, toman los valores predeterminados `inicio=0`, `fin=tamaño de la dimensión`, `paso=1`.

### Subarrays

````python
x = np.arange(10)
x
````

<div class="alert alert-block alert-info">
<b>Pregunta:</b> ¿Qué hace arange?
</div>

````python
x[:5]  # primeros cinco elementos
````

````python
x[5:]  # elementos desde el índice 5 hasta el final
````

````python
x[4:7]  # subarray central del 4 al 7
````

````python
x[::2]  # un elemento de cada dos
````

````python
x[1::2]  # un elemento de cada dos, empezando en el índice 1
````

Un caso potencialmente confuso es cuando el valor de `paso` es negativo.

En general, cuando vemos un índice negativo nos parece un jaleo, pero una vez lo dominas, es muy práctico. Eso sí, hay que usarlos muchos para acostumbrarse. _En mi caso prácticamente los he aprendido de memoria porqué nunca me he llegado a acostumbrar._

En el caso de abajo, vamos a generar un array alrevés. Una vez lo entiendes, se convierte en una forma muy elegante de invertir un array, pero <span style="color:Gold">cuesta acostumbrarse</span>:

````python
x[::-1]  # todos los elementos, invertidos
````

### Subarrays como vistas sin copia

<div class="alert alert-block alert-danger">
<b>Cuidado:</b> Esta parte es fundamental, aseguraos de que os quede claro. 
</div>

Una cosa importante —y extremadamente útil— que hay que saber sobre los slices de arrays es que <span style="color:Gold">devuelven *vistas* en lugar de *copias*</span> de los datos del array. Esta es una de las áreas en las que el slicing de arrays NumPy difiere del slicing de listas de Python: en las listas, `los slices son copias`. Veamos nuestro array de antes:

````python
print(x)
````

Creemos otro pero alreves:

````python
x_alreves = x[::-1]
print(x_alreves)
````

Ahora, si modificamos este array, ¡veremos que el array original también cambia! Observa:

````python
x_alreves[0] = 1337
print(x)
````

Y por supuesto, el que has cambiado también:

````python
print(x_alreves)
````

<div class="alert alert-block alert-success">
<b>Nota:</b> Al principio, me costó creer que esto fuera así, era la primera vez que veía este tipo de dependencia en listas.
</div>

Pero, tiene su sentido.

Este comportamiento predeterminado es realmente útil: significa que <span style="color:Gold">cuando trabajamos con grandes conjuntos de datos</span>, podemos acceder y procesar partes de esos conjuntos sin necesidad de copiar el búfer de datos subyacente.

<div class="alert alert-block alert-info">
<b>Pregunta:</b> ¿Qué peligros implica este comportamiento?
</div>

### Creación de copias de arrays

A veces es útil copiar explícitamente los datos dentro de un array. Esto se puede hacer fácilmente con el método `copy()`:

````python
x_copia = x.copy()
print(x_copia)
````

Si ahora modificamos este array, el array original no se ve afectado:

````python
x_copia[9] = 999
print(x_copia)
````

````python
print(x)
````

## Concatenación y división de Arrays

Todas las funciones anteriores trabajaban con arrays individuales. También es posible combinar varios arrays en uno, y a la inversa, dividir un array en varios.

### Concatenación de arrays

La concatenación, o unión de dos arrays en NumPy, se realiza principalmente usando las funciones `np.concatenate`, `np.vstack` y `np.hstack`.

La más habitual es `np.concatenate`, que toma una tupla o lista de arrays como primer argumento:

````python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np.concatenate([x, y])
````

También puedes concatenar más de dos arrays a la vez:

````python
z = [99, 99, 99]
print(np.concatenate([x, y, z]))
````

### División de arrays

La operación opuesta a la concatenación es la división, implementada por las funciones `np.split`, `np.hsplit` y `np.vsplit`. Para cada una de ellas, podemos pasar una lista de índices que indican los puntos de división:

````python
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)
````
