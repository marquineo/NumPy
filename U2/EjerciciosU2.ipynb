1. Crea un array de NumPy a partir de la lista [1, 2, 3, 4, 5] y verifica su tipo de
dato usando el atributo dtype.

```python
array = np.array([1,2,3,4,5])
print(array.dtype)
```

2. Crea un array con np.array([3.14, 4, 2, 3]) y explica qué sucede
con los tipos de datos

```python
array = np.array([3.14,4,2,3])
print(array.dtype)
#Esto pasa por que si los tipo no coinciden, Numpy realiza una promocion
```

3. Genera un array de 15 elementos relleno de ceros y fuerza que sea de tipo
int32.

```python
array = np.zeros([15],dtype=np.int32)
print(array)
print(array.dtype)
```

```python
np.linspace(0,10,8)
```

7. Crea un array de 10 números aleatorios con distribución uniforme entre 0 y
1, y calcula su media usando .mean().

```python
array = np.random.random(10)
print(array)
array.mean()
```

8. Genera una matriz 5x5 con valores aleatorios de distribución normal

```python
np.random.seed(42)
np.random.normal(0,1,(5,5))
```
